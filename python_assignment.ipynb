```python
                                                             # LIST functions
list1 = ['sachin','lareb','sarath','john cena']

# .append() - used to add some element at the end of the list
list1.append('New element')

# .copy() - makes a copy of the given list
copylist1 =  list1.copy()
copylist1

# .index() - finds out the index of the first matched item
list1.index('sarath')

# .count() - counts the number of occurences of a given arguement
list1.count('sachin')

# .extend() - accepts a new list and merges the elements of the new list to the old one
l2 = [1,2,3,3]
list1.extend(l2)
list1

# .insert() - it is used to add a particular element at a specific position
list1.insert(2,'inserted element at 2')
list1

# .pop() - it is used to remove an element from the end and return it, it doesnt change the original list
list1.pop()

# .remove() -it is used to remove a given element, it all matches will be removed
# original list is changed
list1.remove(3)
list1
                                                            
#  .reverse() it is used to reverse the original list, changes the original list
list1.reverse()
list1

# .reveresed() - it is used to reverse any iterable object not just list, this returns an iterable object
reversed(list1)
list1

''' reverse() is a method of the list class, while reversed() is a built-in function. 
This means that reverse() can only be used on lists, while reversed() can be used on any iterable object.
reverse() modifies the original list in-place, while reversed() returns a
new iterator object that can be used to iterate over the elements of the original list in reverse order. '''

# this is used to sort the data, sort changes the original object and doesnt return anything
# sorted returns a sorted iterable and doesnt change the original list
# list1.sort()
list1

# .clear() -clear function i sused to clear teh given list
list1.clear()
list1
```

```python
                                              ''' DICTIONARY FUNCTIONS '''
dic1 = {'A':10,'B':22,'C':31}                          
dic1

# .copy() - creates a copy of the given dictionary
dic2= dic1.copy()
dic2

# .fromkeys() - it takes a sequence of keys and uses only those keys to make a new dictionary
# and take accepts a default value and asisgns them to new dic
dic3 = dic2.fromkeys(['A','B'],0)
dic3

# .get() - it returns the associated value with the given key,if key isnt there default value is returned
dic1.get("A",'Not exists in the dic')

# .items() - returns the keyvalue pairs of the dic in the form of a list containing tuple
dic1.items()

# .keys() - return a list of all the keys that are present in the list
dic1.keys()

# .values() - returns a list of all the values of the dic
dic1.values()

# .pop() - removes the item with the specified key from the dictionary and
# returns its value. If the key does not exist, a KeyError exception is raised.
# original dic is changed
dic1.pop('B')
dic1

# .popitem() -  removes the item with the specified key from the dictionary and returns its value. 
# If the key does not exist, a KeyError exception is raised.
# also changes the original dic
dic1.popitem()
dic1

# .setdefault() -The setdefault() function sets the value of the specified key to the default 
# value if the key does not exist. If the key already exists, the value is not changed.

dic1.setdefault("B",301)
dic1

# .update()- function updates the dictionary with the specified key-value pairs. 
# If a key already exists in the dictionary, its value is overwritten.
# we can pass a new dic to it or a list of key-value in tuples

dic1.update([('C',101),('D',19)])
dic1

# .clear - it is used to clear the original dictionary
dic1.clear()
dic1
```

```python

                                                          ''' SET FUNCTIONSS '''
set1 = {1,2,3,4,4,5,5,1}
set1
# .add() - it adds an element to the set
set1.add(9)
set1

# .copy() - it makes a shallow copy of the set
set2 = set1.copy()
set2.add(144)

# .difference() - function returns a new set that contains the elements of the set that are not in the specified set.
set2.difference(set1)

# .difference_update() - function removes the elements of the set that are in the specified set.
set2.difference_update(set1)

# intersection() - function returns a new set that contains the elements that are in both sets
set1.intersection(set2)

# intersection_update()- function removes the elements from the set that are not in the specified set.
set1.intersection_update(set2)


# isdisjoint()- function returns True if the sets have no elements in common
set1.isdisjoint(set2)

# issubset() function returns True if all of the elements of the set are in the specified set
set1.issubset(set2)


# issubset() function returns True if all of the elements of the set are in the specified set
set1.issuperset(set2)

# pop() function removes and returns a random element from the set.
set2.pop()

set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

# .discard()- function removes an element from the set if it is present.
set1.discard(9)
set1

# The main difference between discard() and remove() on sets is that discard() does not 
# raise an error if the element to be removed is not present in the set, while remove() does.

# remove() function removes an element from the set if it is present.
set1.remove(2)
set1

# symmetric_difference() function returns a new set that contains the elements that are in either set but not both sets.
set3 = set1.symmetric_difference(set2)
set3

# symmetric_difference_update() function removes the elements from the set that are in the specified set and 
# adds the elements that are in the specified set but not in the set.
set1.symmetric_difference_update(set2)
set1

# union() function returns a new set that contains the elements of both sets.
set1.union(set2)

# update() function adds the elements of the specified set to the set.
set1.update(set2)
set1

# clear() function removes all elements from the set.
set3
set3.clear()
set3
```

```python
                                                        ''' STRING FUNCTIONS '''

strr =  'Hello world! I am a stringg'

# capitalize()
# The capitalize() function converts the first character of a string to uppercase.
strr.capitalize()

# casefold()
# The casefold() function converts a string to lowercase, regardless of the case of the characters.
strr.casefold()

# center()
# The center() function returns a centered string of a specified length.
strr.center(50,"*")

# count()
# The count() function returns the number of times a specified substring occurs in a string.
strr.count("l")

# encode()
# The encode() function returns an encoded version of a string.
strr.encode('ascii')

# endswith()
# The endswith() function returns True if a string ends with a specified substring.
strr.endswith('h')


# expandtabs()
# The expandtabs() function replaces tabs in a string with spaces.
strr.expandtabs()

# find()
# The find() function returns the index of the first occurrence of a specified substring in a string.
strr.find("g")

# format()
# The format() function formats a value according to a format string.

# format_map()
# The format_map() function formats a value according to a format string and a mapping of variable names to value, it accepts dictionary as an arg.

# index()
# The index() function returns the index of the first occurrence of a specified substring in a string.
strr.index("g")

# isalnum()
# The isalnum() function returns True if a string contains only alphanumeric characters.
strr.isalnum()

# isalpha()
# The isalpha() function returns True if a string contains only alphabetic characters.
strr.isalpha()

# isascii()
# The isascii() function returns True if a string contains only ASCII characters.
strr.isascii()

# isdecimal()
# The isdecimal() function returns True if a string contains only decimal characters.
strr.isdecimal()

# isdigit()
# The isdigit() function returns True if a string contains only decimal digits.
strr.isdigit()

# isidentifier()
# The isidentifier() function returns True if a string is a valid identifier.
# valid identifier is something that starts with alphabets can contain numbers, and underscores only.
strr.isidentifier()


# islower()
# The islower() function returns True if a string contains only lowercase characters.
strr.islower()

# isnumeric()
# The isnumeric() function returns True if a string contains only numeric characters.
strr.isnumeric()

# isprintable()
# The isprintable() function returns True if a string contains only printable characters.
strr.isprintable()

# isspace()
# The isspace() function returns True if a string contains only whitespace characters.
strr.isspace()

# istitle()
# The istitle() function returns True if a string is title cased.
strr.istitle()

# isupper()
# The isupper() function returns True if a string contains only uppercase characters.
strr.isupper()

# join()
# The join() function joins a sequence of strings with a specified delimiter.
"*".join(strr)

# ljust()
# The ljust() function returns a left-justified string of a specified length.
strr.ljust(50,'*')

# lower()
# The lower() function converts a string to lowercase.
strr.lower()

# lstrip()
# The lstrip() function removes whitespace characters or a specified string from the left side of a string.
strr.lstrip("H")

# maketrans()
# The maketrans() function creates a translation table.

# partition()
# The partition() function returns a tuple containing the prefix, the substring that matches the delimiter, and the suffix of a string.
#  is used to split a string into three parts based on a specified separato
strr.partition(" ")

# replace()
# The replace() function replaces all occurrences of a specified substring with a new substring.
strr.replace("H",'X')

# rfind()
# The rfind() function returns the index of the last occurrence of a specified substring in a string.
strr.rfind("l")

# rindex()
# The rindex() function returns the index of the last occurrence of a specified substring in a string.
strr.rindex("l")

# rjust()
# The rjust() function returns a right-justified string of a specified length.
strr.rjust(50,"*")

# rpartition()
# The rpartition() method in Python is similar to the partition() method, but it splits the 
# string based on the last occurrence of a specified separator instead of the first occurrence.
strr.rpartition(" ")

# rsplit()
# The rsplit() method in Python is used to split a string into a list of substrings,
#  starting from the rightmost occurrence of a specified separator.
# string.rsplit(separator, maxsplit)
strr.rsplit(" ",2)


# rstrip()
# The rstrip() function removes whitespace characters from the right side of a string.
strr.rstrip("g")

# split()
# The split() function splits a string at a specified delimiter and returns a list of the resulting substrings
strr.split(" ")
```
